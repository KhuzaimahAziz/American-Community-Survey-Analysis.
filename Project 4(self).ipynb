```python
jobs_data.dropna(axis=0, inplace=True)
```

```python
jobs_data['Major_category'].unique()
```

```python
dominant_major=jobs_data.groupby('Major_category')[['Men','Women']].mean()
```

```python
dominant_major
```

```python
dominant_major.plot.barh(figsize=(8,6),color={'Men':'k','Women':'blue'})
```

```python
def dom(row):
    men=row['Men']
    women=row['Women']
    if men > women:
        return 'Men'
    else:
        return 'Women'
```

```python
dominant_major.apply(dom,axis=1)
```

```python
dominant_major['Dom_gens']=dominant_major.apply(dom,axis=1)
```

```python
gen_group=dominant_major.groupby('Dom_gens')
```

```python
gen_group.groups
```

```python
men_grp=gen_group.get_group('Men')
```

```python
men_grp
```

```python
women_grp=gen_group.get_group('Women')
```
